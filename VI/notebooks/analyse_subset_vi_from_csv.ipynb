```python
import pandas as pd

import os
import re

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"# CIFAR10 standard (i.e., in distribution)
```

```python
colors = sns.color_palette("colorblind")
```

```python
sns.palplot(colors)
```

```python
plot_colors = [colors[-3], colors[0], colors[1], colors[1], colors[1]]
```

```python
plt.figure(figsize=(5.5, 5), dpi=300)
sns.set_theme(style="whitegrid", font="Times New Roman")
all_results_df = pd.read_csv("cifar10_all_results_df.csv")

plt.subplot(321)
sns.pointplot(x="model_name", y="test/accuracy",
                 data=all_results_df.reset_index(), palette=plot_colors, ci="sd",
             order=["Deterministic", "All Stoc", "Stoc Input", "Stoc Output", "Last 2 Stoc"], scale=0.5, errwidth=2)
plt.ylabel("Accuracy (%) ($\\uparrow$)", fontsize=8)
plt.gca().set_xticks(0.5+np.arange(all_results_df["model_name"].unique().size-1), minor=True)
plt.xticks(np.arange(all_results_df["model_name"].unique().size-1), ["Deterministic", "Fully\nStochastic", "Input\nLayer\nStochastic", "Output\nLayer\nStochastic", "Input and\nOutput\nLayer\nStochastic"])
plt.grid(which="minor", axis="x", alpha=0.5)
plt.xticks(fontsize=6)
plt.xlabel("")
plt.yticks(fontsize=8)
plt.title("CIFAR-10", fontsize=9, fontweight="bold")

for label_i, label in enumerate(plt.gca().get_xticklabels()):
    label.set_color(plot_colors[label_i])

plt.subplot(323)
sns.pointplot(x="model_name", y="test/negative_log_likelihood",
                 data=all_results_df.reset_index(), palette=plot_colors, ci="sd",
             order=["Deterministic", "All Stoc", "Stoc Input", "Stoc Output", "Last 2 Stoc"], scale=0.5, errwidth=2)
plt.ylabel("NLL ($\\downarrow$)", fontsize=8)
plt.gca().set_xticks(0.5+np.arange(all_results_df["model_name"].unique().size-1), minor=True)
plt.xticks(np.arange(all_results_df["model_name"].unique().size-1), ["Deterministic", "Fully\nStochastic", "Input\nLayer\nStochastic", "Output\nLayer\nStochastic", "Input and\nOutput\nLayer\nStochastic"])
plt.grid(which="minor", axis="x", alpha=0.5)
plt.xticks(fontsize=6)
plt.xlabel("")
plt.yticks(fontsize=8)
for label_i, label in enumerate(plt.gca().get_xticklabels()):
    label.set_color(plot_colors[label_i])


plt.subplot(325)
sns.pointplot(x="model_name", y="test/ece",
                 data=all_results_df.reset_index(), palette=plot_colors, ci="sd",
             order=["Deterministic", "All Stoc", "Stoc Input", "Stoc Output", "Last 2 Stoc"], scale=0.5, errwidth=2)
plt.ylabel("ECE (%) ($\\downarrow$)", fontsize=8)

plt.gca().set_xticks(0.5+np.arange(all_results_df["model_name"].unique().size-1), minor=True)
plt.xticks(np.arange(all_results_df["model_name"].unique().size-1), ["Deterministic", "Fully\nStochastic", "Input\nLayer\nStochastic", "Output\nLayer\nStochastic", "Input and\nOutput\nLayer\nStochastic"])
plt.grid(which="minor", axis="x", alpha=0.5)
plt.xticks(fontsize=6)
plt.xlabel("")
plt.yticks(fontsize=8)
for label_i, label in enumerate(plt.gca().get_xticklabels()):
    label.set_color(plot_colors[label_i])


all_results_df = pd.read_csv("cifar100_all_results_df.csv")
plt.subplot(322)
sns.pointplot(x="model_name", y="test/accuracy",
                 data=all_results_df.reset_index(), palette=plot_colors, ci="sd",
             order=["Deterministic", "All Stoc", "Stoc Input", "Stoc Output", "Last 2 Stoc"], scale=0.5, errwidth=2)
plt.ylabel(None, fontsize=8)
plt.gca().set_xticks(0.5+np.arange(all_results_df["model_name"].unique().size-1), minor=True)
plt.xticks(np.arange(all_results_df["model_name"].unique().size-1), ["Deterministic", "Fully\nStochastic", "Input\nLayer\nStochastic", "Output\nLayer\nStochastic", "Input and\nOutput\nLayer\nStochastic"])
plt.grid(which="minor", axis="x", alpha=0.5)
plt.xticks(fontsize=6)
plt.xlabel("")
plt.yticks(fontsize=8)
plt.title("CIFAR-100", fontsize=9, fontweight="bold")
for label_i, label in enumerate(plt.gca().get_xticklabels()):
    label.set_color(plot_colors[label_i])

plt.subplot(324)
sns.pointplot(x="model_name", y="test/negative_log_likelihood",
                 data=all_results_df.reset_index(), palette=plot_colors, ci="sd",
             order=["Deterministic", "All Stoc", "Stoc Input", "Stoc Output", "Last 2 Stoc"], scale=0.5, errwidth=2)
plt.ylabel(None, fontsize=8)
plt.gca().set_xticks(0.5+np.arange(all_results_df["model_name"].unique().size-1), minor=True)
plt.xticks(np.arange(all_results_df["model_name"].unique().size-1), ["Deterministic", "Fully\nStochastic", "Input\nLayer\nStochastic", "Output\nLayer\nStochastic", "Input and\nOutput\nLayer\nStochastic"])
plt.grid(which="minor", axis="x", alpha=0.5)
plt.xticks(fontsize=6)
plt.xlabel("")
plt.yticks(fontsize=8)
for label_i, label in enumerate(plt.gca().get_xticklabels()):
    label.set_color(plot_colors[label_i])


plt.subplot(326)
sns.pointplot(x="model_name", y="test/ece",
                 data=all_results_df.reset_index(), palette=plot_colors, ci="sd",
             order=["Deterministic", "All Stoc", "Stoc Input", "Stoc Output", "Last 2 Stoc"], scale=0.5, errwidth=2)
plt.ylabel(None)
plt.gca().set_xticks(0.5+np.arange(all_results_df["model_name"].unique().size-1), minor=True)
plt.xticks(np.arange(all_results_df["model_name"].unique().size-1), ["Deterministic", "Fully\nStochastic", "Input\nLayer\nStochastic", "Output\nLayer\nStochastic", "Input and\nOutput\nLayer\nStochastic"])
plt.grid(which="minor", axis="x", alpha=0.5)
plt.xticks(fontsize=6)
plt.xlabel("")
plt.yticks(fontsize=8)
for label_i, label in enumerate(plt.gca().get_xticklabels()):
    label.set_color(plot_colors[label_i])

plt.tight_layout(w_pad=0.5)
plt.savefig("vi_results_variation.pdf", bbox_inches="tight")
```
